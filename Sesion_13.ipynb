```python
import yfinance as yf 
```

```python
ticker = 'AMZN.MX'
```

```python
amazon = yf.download('AMZN.MX',
                    start = '2019-01-01',
                    end = '2023-03-17',
                    progress = False)
                    
```

```python
amazon.head()
```

```python
amazon = amazon[['Adj Close']]
```

```python
amazon.head()
```

```python
amazon = amazon.rename(columns={'Adj Close':'precio_cierre'})
amazon.head()
```

```python
amazon['MA-10'] = amazon['precio_cierre'].rolling(window=10).mean()
```

```python
amazon['precio_cierre'].plot()
plt.title('Precio de la acciÃ³n en Amazon\n 2019-2023')
plt.ylabel('pesos MXN')
plt.show()
```

```python
amazon['MA-10'] = amazon['precio_cierre'].rolling(window=10).mean()
amazon['MA-60'] = amazon['precio_cierre'].rolling(window=60).mean()
```

```python
amazon[['MA-10', 'MA-60']].plot()
plt.title('SeÃ±al de Largo Plazo vs Corto Plazo')
plt.ylabel('pesos MXN')
plt.show()
```
