```python
mun_nl = pd.DataFrame({'municipio':['Monterrey','Guadalupe','Apodaca','San Nicolás','Santiago'],'nacimientos':[500,630,768,400,250]})
```

```python
mun_nl
```

```python
plt.bar(mun_nl['municipio'],mun_nl['nacimientos'])
plt.show()
```

```python
mun_nl = mun_nl.sort_values('nacimientos',ascending=False)
plt.bar(mun_nl['municipio'],mun_nl['nacimientos'],color=('blue','green','red','yellow','grey'))
plt.title('Nacimientos por Municipio',color=('purple'))
plt.xlabel('Municipios de NL',color=('orange'))
plt.ylabel('Cantidad de Nacimientos',color=('brown'))
plt.yticks([0, 200, 400, 600, 800]);
```

```python
fig, (panel1, panel2)=plt.subplots(2,1)
fig.tight_layout()

x1=x
y1=y
panel1.plot(x,y, label='nombre de línea')
panel1.title.set_text('Grafica }1')

x2=mun_nl['municipio']
y2=mun_nl['nacimientos']
panel2.bar(x2,y2)
panel2.title.set_text('Gráfica 2')
```

```python
mun_nl = mun_nl.sort_values('nacimientos',ascending=True)
plt.scatter(mun_nl['municipio'],mun_nl['nacimientos'],color=('blue','green','red','yellow','grey'))
plt.title('Nacimientos por Municipio',color=('purple'))
plt.xlabel('Municipios de NL',color=('orange'))
plt.ylabel('Cantidad de Nacimientos',color=('brown'))
plt.yticks([0, 200, 400, 600, 800]);
```

```python
mun_nl = mun_nl.sort_values('nacimientos',ascending=True)
plt.barh(mun_nl['municipio'],mun_nl['nacimientos'],color=('blue','green','red','yellow','grey'))
plt.title('Nacimientos por Municipio',color=('purple'))
plt.ylabel('Municipios de NL',color=('orange'))
plt.xlabel('Cantidad de Nacimientos',color=('brown'))
plt.xticks([0, 200, 400, 600, 800]);
```
